```python
def convert(org,idx='0'):
    data = {}
    marathi_name = org['@TEXT'].strip()
    data['marathi_name'] = marathi_name
    data['english_name'] = translator.translate(marathi_name).text
    data['id'] = idx
    
    print(marathi_name)
    if(org.get('node')):
        childrens = org['node']
        if type(childrens) == dict:
            childrens = [childrens]
            
        d_child = []
        
        ids = idx
        count = 0 
        
        for child in childrens:
            d = convert(child,ids+'-'+str(count))
            count+=1
            d_child.append(d)
        
        data['children'] = d_child 

    
    return data
    
```

```python
data = convert(root)
```
